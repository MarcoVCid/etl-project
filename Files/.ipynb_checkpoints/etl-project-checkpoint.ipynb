```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
top50 = "../Resources/top50.csv"
top50_df = pd.read_csv(top50)
top50_df.head(3)
```

```python
spotify = "../Resources/1921 - 2020/data.csv"
spotify_df = pd.read_csv(spotify)
spotify_df.head(3)
```

### Transform Top 50 DataFrame

```python
print(top50_df.columns)
print(spotify_df.columns)
```

```python
df_top50 = top50_df.rename(columns={
                        'id':'id',
                        'Track.Name' : 'song_name',
                        'Artist.Name':'artist',
                        'Energy':'energy',
                        'Danceability':'danceability',
                        'Loudness..dB..':'loudness',
                        'Valence.':'valence',
                        'Length.':'song_length',
                        'Acousticness':'acousticness',
                        'Speechiness.':'speechness',
                        'Popularity':'popularity'
                        
                        })
```

```python
df_top50 = df_top50[['id', 'song_name', 'artist', 'energy', 'danceability', 'loudness', 'valence', 'song_length',
                   'acousticness', 'speechness', 'popularity']]
df_top50.head()
```

### Transform 1921-2020 DataFrame

```python
print(spotify_df.columns)
```

```python
df_spotify = spotify_df.rename(columns={
                        'id':'id',
                        'name' : 'song_name',
                        'artists':'artist',
                        'energy':'energy',
                        'danceability':'danceability',
                        'loudness':'loudness',
                        'valence':'valence',
                        'duration_ms':'song_length',
                        'acousticness':'acousticness',
                        'speechiness':'speechness',
                        'popularity':'popularity'
                        
                        })
```

```python
df_spotify = df_spotify[['id', 'song_name', 'artist', 'energy','danceability', 'loudness', 'valence', 'song_length',
                   'acousticness', 'speechness', 'popularity']]
df_spotify.head()
```

```python
df_top50.head()
```

```python
df_spotify['artist'] = df_spotify['artist'].str.replace("[", "").str.replace("]", "")
df_spotify['artist'] = df_spotify['artist'].str.replace("'", "")
```

```python
df_spotify['energy'] = round(df_spotify['energy'] * 100, 0).astype(int)
```

```python
df_spotify['danceability'] = round(df_spotify['danceability'] * 100, 0).astype(int)
```

```python
df_spotify['loudness'] = round(df_spotify['loudness'], 0).astype(int)
```

```python
df_spotify['valence'] = round(df_spotify['valence'] * 100, 0).astype(int)
```

```python
df_spotify['song_length'] = round(df_spotify['song_length'] / 1000, 0).astype(int)
```

```python
df_spotify['acousticness'] = round(df_spotify['acousticness'] * 100, 0).astype(int)
```

```python
df_spotify['speechness'] = round(df_spotify['speechness'] * 100, 0).astype(int)
```
